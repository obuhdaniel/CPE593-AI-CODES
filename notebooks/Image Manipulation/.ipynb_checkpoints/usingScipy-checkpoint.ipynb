```python
from scipy import misc
import imageio.v2 as imageio
import matplotlib.pyplot as plt
```

```python
#Lets read the image using imageio
f = imageio.imread('racoon.png')
print(f.dtype, f.shape)
plt.imshow(f)
plt.show()
```

```python
#lets tint the image
import numpy as np
tint_array = np.array([1, 0.45, 0.3])
f_tint = f * tint_array
#Notice here the need to clip the values of the image here
f_tint = np.clip(f_tint, 0, 255).astype(np.uint8)

imageio.imsave('racoon_tint.png',f_tint)
print(f_tint.dtype, f_tint.shape)
plt.imshow(f_tint)
plt.show()
```
