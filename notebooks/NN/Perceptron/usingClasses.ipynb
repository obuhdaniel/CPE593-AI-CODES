```python
import numpy as np

class NeuralNetwork:
    def __init__(self):
        pass
        
    
    n_input = 3
    n_hidden = 2
    n_output = 1

    learning_rate = 0.1
    n_iterations = 100


    w1 = np.random.rand(n_input, n_hidden)
    b1 = np.random.rand(n_hidden)
    w2 = np.random.rand(n_hidden, n_output)
    b2 = np.random.rand(n_output)

    def print_weights_and_biases(self):
        # Print the initialized weights and biases
        print("Weights and biases initialized:")
        print("w1:", self.w1)
        print("b1:", self.b1)
        print("w2:", self.w2)
        print("b2:", self.b2)



    # Create test and train dataset
    X_train = np.random.rand(80, n_input)
    y_train = np.random.rand(80, n_output)
    X_test = np.random.rand(20, n_input)
    y_test = np.random.rand(20, n_output)





    def sigmoid(self, z):
        return 1.0 / (1.0 + np.exp(-z))

    def sigmoid_derivative(self, z):
        return z * (1 - z)
    
    def relu(self, z):
        return np.maximum(0, z)
    
    def relu_derivative(self, z):
        return np.where(z > 0, 1, 0)
    
  

    def feedforward(self):

        hidden_layer = self.relu(np.dot(self.X_train, self.w1) + self.b1)
        output_layer = self.sigmoid(np.dot(hidden_layer, self.w2) + self.b2)
        return hidden_layer, output_layer
        

    def backpropagate(self):
        # Backpropagation algorithm to update weights and biases
        # Calculate the error at the output layer
        output_error = self.y_train - self.output_layer
        output_delta = output_error * self.sigmoid_derivative(self.output_layer)
        hidden_layer_error = output_delta.dot(self.w2.T)
        hidden_layer_delta = hidden_layer_error * self.relu_derivative(self.hidden_layer)
        # Update weights and biases
        self.w2 += self.hidden_layer.T.dot(output_delta) * self.learning_rate
        self.b2 += np.sum(output_delta, axis=0) * self.learning_rate
        self.w1 += self.X_train.T.dot(hidden_layer_delta) * self.learning_rate
        self.b1 += np.sum(hidden_layer_delta, axis=0) * self.learning_rate



  

    def loss(self):
        # Calculate the loss using Mean Squared Error
        for i in range(self.n_iterations):
            self.hidden_layer, self.output_layer = self.feedforward()
            self.backpropagate()
            #print weights and biases after each iteration
            print("Iteration:", i+1)
            self.print_weights_and_biases()
        
        return np.mean(np.square(self.y_train - self.output_layer))
    
    
    def predict(self, x):
        x = np.array(x).reshape(1, -1)
        hidden_layer = self.relu(np.dot(x, self.w1) + self.b1)
        output_layer = self.sigmoid(np.dot(hidden_layer, self.w2) + self.b2)
        return output_layer
    



if __name__ == "__main__":
    nn = NeuralNetwork()

    x = [7, 7, 7]  # Ensure the input has the correct shape (1, 3)
    
    nn.print_weights_and_biases()
    loss = nn.loss()
    print("Loss:", loss)
    prediction = nn.predict(x)
    print("Prediction:", prediction)
    print("Weights and biases after training:")
    nn.print_weights_and_biases()

```
